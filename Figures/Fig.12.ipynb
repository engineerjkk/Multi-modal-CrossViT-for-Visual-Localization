```python
# Precision@K 그래프를 그리기 위한 코드 재시도
import matplotlib.pyplot as plt
# 데이터 정의
ks = [1, 5, 20, 50, 100, 150, 200]

# 5th 학생 및 NetVLAD
precision_5th_student = [0.9016, 0.8407, 0.7473, 0.6924, 0.7150, 0.7618, 0.8085]
precision_5th_netvlad = [0.9433, 0.8198, 0.6023, 0.4594, 0.4249, 0.4360, 0.4624]

# 4th 학생 및 NetVLAD
precision_4th_student = [0.8843, 0.8222, 0.7314, 0.6843, 0.7144, 0.7672, 0.8152]
precision_4th_netvlad = [0.9306, 0.8165, 0.5954, 0.4540, 0.4185, 0.4311, 0.4575]

# 3rd 학생 및 NetVLAD
precision_3rd_student = [0.9063, 0.8426, 0.7371, 0.6871, 0.7258, 0.7806, 0.8302]
precision_3rd_netvlad = [0.9595, 0.8352, 0.6133, 0.4780, 0.4462, 0.4545, 0.4780]

# 2nd 학생 및 NetVLAD
precision_2nd_student = [0.8935, 0.8425, 0.7471, 0.6963, 0.7302, 0.7827, 0.8305]
precision_2nd_netvlad = [0.9363, 0.8144, 0.5954, 0.4512, 0.4175, 0.4530, 0.4530]

# 1st 학생 및 NetVLAD
precision_1st_student = [0.9021, 0.8360, 0.7284, 0.6845, 0.7176, 0.7714, 0.8199]
precision_1st_netvlad = [0.9435, 0.8216, 0.5880, 0.4536, 0.4206, 0.4322, 0.4560]

plt.figure(figsize=(12, 9))

# 학생 모델들에 대한 그래프 그리기
plt.plot(ks, precision_5th_student, marker='o', linestyle='-', color='blue', label='5th Student')
plt.plot(ks, precision_4th_student, marker='x', linestyle='-', color='lightblue', label='4th Student')
plt.plot(ks, precision_3rd_student, marker='^', linestyle='-', color='skyblue', label='3rd Student (Max)')
plt.plot(ks, precision_2nd_student, marker='s', linestyle='-', color='deepskyblue', label='2nd Student')
plt.plot(ks, precision_1st_student, marker='*', linestyle='-', color='dodgerblue', label='1st Student')

# NetVLAD 모델들에 대한 그래프 그리기
plt.plot(ks, precision_5th_netvlad, marker='o', linestyle='--', color='red', label='5th NetVLAD')
plt.plot(ks, precision_4th_netvlad, marker='x', linestyle='--', color='salmon', label='4th NetVLAD')
plt.plot(ks, precision_3rd_netvlad, marker='^', linestyle='--', color='lightcoral', label='3rd NetVLAD')
plt.plot(ks, precision_2nd_netvlad, marker='s', linestyle='--', color='firebrick', label='2nd NetVLAD')
plt.plot(ks, precision_1st_netvlad, marker='*', linestyle='--', color='darkred', label='1st NetVLAD')

plt.xlabel('K')
plt.ylabel('Precision@K')
plt.title('Precision@K Curve')
plt.gca().invert_xaxis()
#plt.legend()
plt.grid(True)
plt.show()
```
